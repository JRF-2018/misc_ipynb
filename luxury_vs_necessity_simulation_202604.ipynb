# 必需品と贅沢品の宇宙的独立関係に関する数理的考察

## 概要
本稿は、「必需品と贅沢品の配分は、ある条件下では独立に決定されるべきである」という独創的な議論を数理的に定式化する試みである。人類の存続を確率過程としてモデル化し、無限の価値を持つ可能性のある知的財産（贅沢品）の期待価値が不確定になることを示す。これにより、両者の最適な配分比率が決定不能となり、結果として「独立性」が論理的に導かれることを論じる。

---

## 1. 前提条件と定義
議論の前提を以下のように数学的に定義する。

* **財の分類**: 全ての財の集合を $G$ とする。$G$ は、必需品の集合 $B$ (Basics) と贅沢品の集合 $L$ (Luxuries) に直和分割される。

    $$G = B \oplus L \quad (B \cap L = \emptyset)$$

  知的財産（Intellectual Property, $I$）は贅沢品に含まれるとする: $I \subset L$。

* **人類の存続**: 時刻 $t$ における人類の人口を $N(t) \in \mathbb{Z}^+$ とする。人類の目標は「中位存続」、すなわち $N(t) > 0$ を可能な限り長く維持することである。

* **必需品の制約**: 存続のためには、最低限の必需品量 $B_{min}$ が必要である。時刻 $t$ に生産される必需品の量を $b(t)$ とすると、存続の必要条件は以下となる。

    $$b(t) \ge B_{min}$$

---

## 2. 絶滅の確率論的必然性
不確実性下での人類の存続を、マルコフ連鎖のような確率過程として考える。

* 人口 $N(t)$ を状態変数とする。$N(t)=0$ は「絶滅」という吸収状態である。
* 外部からの革新的な介入がない限り、有限の人口 $N(t)$ がランダムな変動（戦争、災害、疫病など）にさらされ続ける場合、十分に長い時間 $T$ の後、絶滅状態に到達する確率は $1$ に収束する。これは「ギャンブラーの破産問題」としても知られる原理である。

    $$\lim_{T \to \infty} P(N(T) = 0 \mid N(0) > 0) = 1$$

  これを絶滅事象 $E_{ext}$ とすると、$P(E_{ext}) = 1$ となる。

---

## 3. 無限の宇宙における確率 0 の生存可能性
議論の核心である「確率 0 でも望みがある」という部分を定式化する。

* **時間概念の拡張**: 物理的時間 $t$ とは別に、人類が経験する概念的時間 $\tau$ を考える。例えば、宇宙の寿命を $T_{univ}$ として、以下の写像 $f$ を定義する。

    $$\tau(t) = \tan\left(\frac{\pi}{2} \cdot \frac{t}{T_{univ}}\right)$$

  この写像により、有限の物理時間区間 $[0, T_{univ})$ が、無限の概念的時間区間 $[0, \infty)$ に対応づけられる。これは、宇宙を実質的に無限の試行回数を持つ存在と見なすことを数学的に表現している。

* **確率 0 事象の実現**: 連続的な確率空間において、ある事象 $A$ の確率が $P(A)=0$ であっても、その事象が不可能（$A = \emptyset$）であることを意味しない。同様に、人類の永続的存続という事象 $S$ の確率が $P(S)=0$ であっても、無限の宇宙（試行空間）の中では、その事象が起こる「経路」が存在しうる。

---

## 4. 知的財産の無限価値と期待値の不確定性
この確率 0 の生存を可能にするのが、贅沢品から生まれる知的財産 $I$ である。

* **知的財産の価値**: 絶滅を回避し、確率 0 の生存 $S$ を実現するような画期的な知的財産 $I^*$ が生まれるとする。この $I^*$ の価値 $V(I^*)$ は、回避される損失（人類の全価値）に等しい。これを絶滅確率の逆数として定義すると、

    $$V(I^\*) \propto \frac{1}{P(\text{survival without } I^\*)} = \frac{1}{0} \to \infty$$

  となり、その価値は無限大と評価される。

* **期待値の計算**: ある期間内にこの無限の価値を持つ知的財産 $I^*$ が生まれる確率を $p_{I^*}$ とする。この確率は極めて小さいと考えられる。

    $$p_{I^\*} \approx 0$$

  このとき、贅沢品にリソースを投下することの期待価値 $E[V_L]$ は、

    $$E[V_L] = p_{I^\*} \times V(I^\*) \approx 0 \times \infty$$

  となり、これは数学的に「不定形 (Indeterminate form)」である。つまり、贅沢品への投資がもたらす長期的なリターンは、計算によって評価することができない。

---

## 5. 結論：配分の独立性
期待値が評価不能であることから、論理的な結論が導かれる。

* **最適化問題**: 全資源 $R$ のうち、割合 $\alpha$ を贅沢品に、$(1-\alpha)$ を必需品に配分する。目標は、人類の存続確率を最大化する $\alpha$ を見つけることである。

    $$\max_{\alpha} \text{SurvivalProbability}(\alpha)$$

  ただし、必需品の制約から $\alpha$ には上限 $\alpha_{max}$ が存在する。

    $$(1-\alpha)R \ge B_{min} \implies \alpha \le 1 - \frac{B_{min}}{R} = \alpha_{max}$$

* **決定不可能性**: 考慮すべき領域 $\alpha \in [0, \alpha_{max}]$ において、$\alpha$ を増やすこと（贅沢品への投資）の価値は $E[V_L]$ に依存するが、この値が不定形であるため、効用関数を具体的に定義できず、最適化問題を解くことができない。

* **独立性**: したがって、必需品の最低ライン $B_{min}$ を確保した上で、どれだけの資源を贅沢品に割り振るか（すなわち $\alpha$ の値をどう設定するか）は、数理的な最適化によっては決まらない。この選択は、他の変数（必需品の量）から「独立」した、社会の価値観や信念に基づく政治的な決定となる。

---

## 6. 補足：国家間競争における非独立性
この独立性の議論は、閉じた系（地球全体）で成り立つ。国家間の競争が存在する場合、状況は変わる。

* **ゲーム理論的状況**: 2つの国（国1, 国2）の配分比率をそれぞれ $\alpha_1, \alpha_2$ とする。国1の成功確率（効用） $U_1$ は、自国の選択だけでなく、相手国の選択にも依存する。

    $$U_1 = f(\alpha_1, \alpha_2)$$

  これはゲーム理論の構造であり、最適な戦略 $\alpha_1^*$ は相手の戦略 $\alpha_2$ の関数となる（最適反応関数）。

    $$\alpha_1^* = g(\alpha_2)$$

  この場合、$\alpha_1$ の決定はもはや独立ではなく、競争環境によって強く制約される。各国は、相手の出方を探りながら自国の配分を調整せざるを得ない。
</blockquote>

# 数学的チェック：「必需品と贅沢品の宇宙的独立関係」

全体的な印象から言うと、**議論の流れはおおむね元の文章を忠実に数式化できています**が、いくつか数学的に問題のある点・注意すべき点があります。順に見ていきます。

---

## ✅ 正しい・問題ない部分

### Section 1（定義）
$G = B \oplus L$ という直和分割は適切です。

### Section 2（絶滅の確率論的必然性）
$$\lim_{T \to \infty} P(N(T) = 0 \mid N(0) > 0) = 1$$
これはギャンブラーの破産問題の一般化として概ね正しいです。ただし、**厳密には条件が必要**（後述）。

### Section 3（確率0事象の実現）
連続確率空間における「確率0 ≠ 不可能」という議論は数学的に正しいです。ルベーグ測度の意味での確率0事象の話として適切。

### Section 5（ゲーム理論）
$U_1 = f(\alpha_1, \alpha_2)$ の構造とナッシュ均衡的な最適反応関数の発想は正しいです。

---

## ⚠️ 問題・修正が必要な部分

### Section 2：ギャンブラーの破産の適用条件が不足

ギャンブラーの破産で確率1の絶滅が保証されるのは、**ランダムウォークの増分の期待値が0以下**のときです。期待値が正（人口が平均的に増える）の場合、確率1の絶滅は成立しません。

元の文章の趣旨（不確実性下では長期的に絶滅する）を数学的に正確にするなら、次を追記すべきです：

> 「増分の期待値が $\mu \leq 0$ であるか、または $N(t)$ が上限のある有界な空間に閉じ込められている場合に限り、$P(E_{ext}) = 1$」

---

### Section 3：時間写像の定式化に問題あり

Geminiが提示した写像：
$$\tau(t) = \tan\left(\frac{\pi}{2} \cdot \frac{t}{T_{univ}}\right)$$

これは $t \to T_{univ}$ で $\tau \to \infty$ となりますが、**$t = 0$ での値が $\tau = 0$** で問題なく、$t \in [0, T_{univ})$ を $[0, \infty)$ に写すという意味では数学的に正しいです。

ただし元の文章にあった $y = \tan(ax)$ や $y = \exp(\tan(ax))$ との対応が変わっています。$\exp(\tan(ax))$ を使うと $t \to T_{univ}$ で二重に発散し、より「急激な」無限化が表現できますが、Geminiはよりシンプルな形にまとめました。これは**簡略化として許容範囲**です。

---

### Section 4：最大の問題点 — $\frac{1}{0}$ の扱い

$$V(I^*) \propto \frac{1}{P(\text{survival without } I^*)} = \frac{1}{0} \to \infty$$

これは**数学的に不正確**です。問題点は二つあります：

**①「比例」関係の根拠がない**
なぜ価値が生存確率の逆数に比例するのか、正当化されていません。

**②$\frac{1}{0}$は極限として扱うべき**
$\frac{1}{0}$ は直接定義できません。正しくは：

$$P(\text{survival}) = \epsilon \to 0^+ \implies V(I^*) \propto \frac{1}{\epsilon} \to +\infty$$

と極限として書くべきです。これは細かいようで、$0 \times \infty$ の不定形を議論するときに重要な違いになります。

---

### Section 4：$0 \times \infty$ の不定形の扱い

$$E[V_L] = p_{I^*} \times V(I^*) \approx 0 \times \infty$$

「不定形 (Indeterminate form)」という結論は**方向性は正しい**のですが、厳密には「$0 \times \infty$」は**極限の文脈でのみ不定形**です。

つまり：
$$\lim_{\epsilon \to 0} \epsilon \cdot \frac{1}{\epsilon^r}$$
は $r$ の値によって $0$、有限値、$\infty$ のいずれにもなりえます。これが「評価できない」という元の文章の意図に対応します。

もう少し丁寧に書くなら：

> $p_{I^*}(\alpha)$ と $V(I^*)$ がそれぞれ $\alpha$ の関数であり、その積の $\alpha \to \alpha_{max}$ における極限が、関数の形に依存して値が定まらないため、最適化が不能になる

とするほうが数学的に誠実です。

## 修正版：Section 4の厳密化

### 知的財産の価値の定式化

$I^*$ が「絶滅を回避する知的財産」だとします。$I^*$ がない場合の存続確率を $\epsilon > 0$、$I^*$ がある場合の存続確率を $q > \epsilon$ とします。

$I^*$ の価値を、**回避された損失の期待値**として定義します：

$$V(I^*) = \frac{q - \epsilon}{\epsilon} \cdot W$$

ここで $W$ は「人類の全価値」（有限の定数）です。$\epsilon \to 0^+$ のとき：

$$V(I^*) \to \infty$$

これで $1/0$ を直接使わずに済みます。

---

### 期待値の不定性の厳密化

贅沢品への配分比率を $\alpha \in [0, \alpha_{max}]$ とします。$I^*$ が生まれる確率 $p(\alpha)$ は $\alpha$ の増加関数で：

$$p(\alpha) \geq 0, \quad p(0) = 0, \quad \lim_{\alpha \to \alpha_{max}} p(\alpha) = p_{max} \approx 0$$

また、$I^*$ の価値を $V(\epsilon)$ と書くと（$\epsilon$ は $\alpha$ に依存しない定数として扱う）：

$$E[V_L](\alpha) = p(\alpha) \cdot V(\epsilon)$$

ここで $p(\alpha) \to 0$、$V(\epsilon) \to \infty$ のとき、この積は：

$$\lim_{\epsilon \to 0} p(\alpha) \cdot V(\epsilon) = \lim_{\epsilon \to 0} p(\alpha) \cdot \frac{q-\epsilon}{\epsilon} \cdot W$$

$p(\alpha)$ と $\epsilon$ の関係が独立に指定されている（$\alpha$ を増やしても $\epsilon$ は外生的に決まる）ため、この極限は $p(\alpha)$ の収束速度と $\epsilon$ の収束速度の比に完全に依存します。

具体的に：

- $p(\alpha) \sim C\epsilon^r$（$r > 0$）と仮定すると：

$$E[V_L] \sim C\epsilon^r \cdot \frac{1}{\epsilon} \cdot W = CW\epsilon^{r-1}$$

- $r > 1$ のとき：$E[V_L] \to 0$（贅沢品に価値なし）
- $r = 1$ のとき：$E[V_L] \to CW$（有限の定数）
- $r < 1$ のとき：$E[V_L] \to \infty$（贅沢品に全賭けすべき）

**$r$ の真の値は原理的に知ることができません。** これが「評価不能」の数学的実体です。

## Section 2の修正：絶滅確率=1の条件

ギャンブラーの破産の適切な条件を明示します。

$N(t)$ を確率過程とし、増分 $\Delta N = N(t+1) - N(t)$ の期待値を $\mu$ とします：

- **$\mu < 0$ のとき**：$P(N(T)=0) = 1$（有限時間で絶滅）
- **$\mu = 0$ のとき**：$P(N(T)=0) = 1$（ランダムウォーク、再帰性）
- **$\mu > 0$ のとき**：$P(N(T)=0) < 1$（正の漂流があれば生存確率あり）

元の議論の前提を保つには：

$$P\left(\exists T : N(T) = 0 \mid N(0) = n\right) = 1 \quad \forall n \in \mathbb{Z}^+$$

### 最適化問題の設定

$$\max_{\alpha \in [0, \alpha_{max}]} F(\alpha), \quad F(\alpha) = E[\text{SurvivalValue}(\alpha)]$$

$$F(\alpha) = \underbrace{f_B(1-\alpha)}_{\text{必需品の寄与：単調増加・有界}} + \underbrace{E[V_L](\alpha)}_{\text{贅沢品の寄与：不定形}}$$

$$F(\alpha) \text{ の最大化問題は well-posed ではない}$$

これを「**配分比率 $\alpha$ が決定不能（underdetermined）**」と言います。「独立」とは、$\alpha$ の最適値が $f_B$ の情報だけからは決定されない、という意味です。

形式的に書けば：

$$\nexists \alpha^* \in [0, \alpha_{max}] \text{ such that } \alpha^* = \arg\max F(\alpha) \text{ is uniquely determined}$$

## 危機の近傍の数理的描写

### 基本設定

時刻 $t$ において、危機 $c$ が発生する確率を考えます。危機までの距離（時間的・社会的）を $d > 0$ として、$d \to 0$ が「危機の近傍」です。

贅沢品の集合を $L = \{l_1, l_2, \ldots, l_n, \ldots\}$（可算無限）とします。危機 $c$ に対して「有効な贅沢品」の集合を：

$$L^*(c) \subset L$$

と書きます。$d$ が大きい（危機が遠い）とき、$L^*(c)$ は**未確定**です。$d \to 0$ のとき $L^*(c)$ が**確定していく**というのが今回描きたい構造です。

---

### $d \to 0$ における「有効集合」の収束

$d$ の関数として、贅沢品 $l_i$ が有効である確率を：

$$p_i(d) = P(l_i \in L^*(c) \mid \text{crisis distance} = d)$$

と定義します。

- $d$ が大きい（危機が遠い）：$p_i(d) \approx \frac{1}{|L|}$（どれが有効かほぼ均等に不明）
- $d \to 0$（危機の近傍）：ある $l_i$ については $p_i(d) \to 1$、他は $p_i(d) \to 0$

これはシャノンエントロピーで測ると：

$$H(d) = -\sum_{i} p_i(d) \log p_i(d)$$

$$H(d \gg 0) \approx \log |L| \quad \text{（最大不確実性）}$$
$$H(d \to 0) \to 0 \quad \text{（確定）}$$

つまり**危機への距離 $d$ はエントロピー $H(d)$ の単調増加関数**として描けます。

---

### 「奇跡」の数理的位置づけ

投資の意思決定は $d = D$（危機がまだ遠い時点）で行われます。そのとき投資家は $l_j$ に投資したとします。

後から見て $l_j \in L^*(c)$ だった確率は：

$$P(l_j \in L^*(c)) = p_j(D) \approx \frac{|L^*(c)|}{|L|}$$

$|L|$ が大きく $|L^*(c)|$ が小さければ、これは非常に小さい値です。

ここで重要なのは、$d = D$ 時点での意思決定者には $H(D)$ が大きく、つまり**情報がほぼない**状態であることです。それにもかかわらず $l_j \in L^*(c)$ に投資していたとすれば：

$$\text{「奇跡」} = \frac{p_j(0) - p_j(D)}{1 - p_j(D)} \to 1 \quad (d \to 0)$$

これは「事後的に見て正しかった投資」の希少性を表します。

---

### 転嫁のダイナミクス

$d \to 0$ の過程で何が起きるか。贅沢品 $l_j$ が有効と確定していく過程を：

$$\frac{dp_j(d)}{dd} < 0 \quad \text{（$d$ が減るほど $p_j$ は確定方向へ動く）}$$

と書けます。$p_j(d) \to 1$ となった瞬間、社会はその $l_j$ に資源を集中投下します。

これが「贅沢品から必需品への転嫁」です。数式で書けば：

$$\alpha_j(d) = \begin{cases} \text{小} & (d \gg 0, \; l_j \text{ はまだ贅沢品}) \\ \to 1 & (d \to 0, \; l_j \text{ が必需品に転嫁}) \end{cases}$$

転嫁の速度は $\frac{d\alpha_j}{dd}$ で、これは：

$$\frac{d\alpha_j}{dd} = -\frac{dp_j}{dd} \cdot \frac{1}{H(d)}$$

$H(d) \to 0$（危機が近づいて不確実性が消える）と $\frac{dp_j}{dd}$ が大きくなるほど、転嫁は急激になります。

---

### 全体像のまとめ

```
d 大（危機遠）          d → 0（危機の近傍）
   
H(d) 大                H(d) → 0
「どれが有効か不明」    「l_j が有効と確定」
贅沢品として分散投資    l_j が必需品に転嫁
α_j 小・均等           α_j → 1（集中）
奇跡の確率 ≈ 0         奇跡だったと判明
```

---

### 独立性への含意

この描写から見えてくるのは：

> $d$ が大きい平時においては、どの贅沢品が有効かわからないため、**$\alpha$ の最適配分は依然として決定不能（独立性が成立）**

> $d \to 0$ の危機近傍では、$L^*(c)$ が確定し始めるため**独立性が崩壊して必需品的論理が支配**する

つまり「独立性」は平時の性質であり、危機の近傍ではその独立性が**位相的に変化**するのです。

これはある種の**相転移**として描けます。$d = d_c$（臨界距離）において：

$$\frac{dH}{dd}\bigg|_{d=d_c} \to -\infty$$

エントロピーが急激に落ちる臨界点が存在し、そこで「贅沢品の論理」から「必需品の論理」への相転移が起きる。これが危機の近傍の数理的な姿だと思います。

## 対応関係の確認

ご直感を整理すると：

$$\text{析出} \leftrightarrow \text{事前学習}$$
$$\text{危機} \leftrightarrow \text{強化学習}$$

これは構造的に非常に正確だと思います。

---

## 析出 ↔ 事前学習

事前学習は大量のデータから**共通構造を抽出**する過程です。

$$\min_{P} H(P) \quad \text{over 大量のサンプル}$$

重なりのある $l_j$ たちから $F_0$（核）を析出するのは、まさにこれです。多様なデータの中から**低エントロピーの共通表現**を見つける。

事前学習後のモデルは「何が似ていて何が違うか」という構造を持っています。析出後の $L/{\sim}$ が有限の同値類を持つのと同じです。

---

## 危機 ↔ 強化学習

強化学習は**報酬信号**によって行動を更新する過程です。

$$\alpha(t+1) = \alpha(t) + \eta \cdot R(t) \cdot \nabla \log \pi(\alpha(t))$$

危機はまさに強烈な報酬信号 $R(t)$ です。

- 危機に正しく対応できた：$R > 0$（有効な贅沢品に投資していた「奇跡」）
- 対応できなかった：$R < 0$（絶滅方向）

そして強化学習の特徴は：

$$\xi_2(d) \to 0$$

事前学習では消えなかったモデル不確実性が、**報酬という外部信号によって初めて解消される**。

これが危機の本質です。平時の学習では得られない信号が来る。

---

## 両者の関係

ただし重要な非対称性があります。

事前学習は：
$$\text{多様なデータ} \to \text{汎用的な表現} \quad (\xi_1 \text{ を減らす})$$

強化学習は：
$$\text{特定の報酬} \to \text{特定の行動に収束} \quad (\xi_2 \text{ を減らす})$$

そして：

$$\text{事前学習なしの強化学習} \implies \text{析出なしの危機対応}$$

これは**表現が貧困なまま報酬信号だけ来る**状態であり、うまく対応できません。贅沢品の厚みがないまま危機が来る状況に対応します。

逆に：

$$\text{強化学習なしの事前学習のみ} \implies \text{危機なしの析出のみ}$$

これは構造は持っているが**報酬信号で更新されない**状態です。平時に贅沢品を蓄積しているが、どれが有効かは永遠にわからない状態に対応します。

---

## 全体像

```
事前学習（析出）
  大量サンプル → 共通構造 F₀ の抽出
  H(L) → H(L/~)（エントロピー低下）
  ξ₁ → 0
  ただし ξ₂ は残る
        ↓
強化学習（危機）
  報酬信号 R(t) が来る
  ξ₂ → 0
  α が収束・安定
  L/~ の中のどれが有効かが確定
```

そしてこの対応が示唆することは：

### 1. 析出 ＝ 事前学習（タスク非依存の表現学習）

平時において、人類が多様な「贅沢品（基礎研究、芸術、思想など）」に投資し、その重なり合いから本質的な概念を「析出」していく過程は、まさにLLMの事前学習（自己教師あり学習）そのものです。

* **サバイバルという報酬:** 例えば「エネルギーを確保できれば生存（$+1$）、できなければ絶滅（$-1$）」という強化学習の環境（マルコフ決定過程）に、社会というエージェントが突然放り込まれます。
* **探索から利用（Exploitation）へのシフト:** エントロピーの議論であった通り、危機の近傍ではランダムな「探索」にリソースを割いている余裕はありません。社会は、生き残るために最も期待値の高い行動を計算し、全リソースをそこに集中（必需品への転化）させます。

### このモデルが証明する「贅沢品の独立性」の真の価値

この「事前学習 $\to$ 強化学習」というパラダイムを用いると、あなたの最初の問いである**「なぜ贅沢品は必需品のバッファ（保険）として従属するのではなく、独立していなければならないのか？」**に対する、最も力強い数理的解答が導かれます。

強化学習（危機時の適応）は、ゼロから始めると**サンプル効率が極めて悪い**（正解に辿り着くまでに何度も失敗を繰り返さなければならない）という致命的な弱点があります。現実の社会において「何度も絶滅して学ぶ」ことは許されません。

社会が未知の危機（全く新しい強化学習タスク）に対して、致命傷を負うことなく「ゼロショット」あるいは「フューショット」で素早く適応できるのはなぜか。
それは、平時にタスク（必需品・軍事・経済的要請）から完全に独立した状態で、膨大で無駄に見える事前学習（贅沢品の析出）を済ませていたからです。

もし、贅沢品が最初から「危機のバッファ」として必需品に従属させられていたらどうなるでしょう。それは、最初から特定のタスク（例：過去の戦争の延長線上の戦術）のためだけに最適化されたモデルを作ることと同義です。その場合、全く想定外の危機（例：未知のウイルスや環境激変）という新しい報酬関数が提示された瞬間、潜在空間にその事象を解釈する次元が存在せず、適応できずに絶滅します。

この点を Gemini さんに Python で例にしてもらったのが次のソースになる。おそらく loss_C がポイントで「なんとなく特定の行動に対応している」ここが「通常社会」の在り方の意味なのだろう。


````python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# シードの固定（再現性のため）
torch.manual_seed(42)
np.random.seed(42)

# ==========================================
# 1. 環境と危機の定義の厳密化
# ==========================================
input_dim = 100
hidden_dim = 20
num_samples = 2000

X = torch.randn(num_samples, input_dim)

# 既知の危機（タスク1）：最初の20次元のみに依存
W_known = torch.zeros(input_dim, 1)
W_known[:20, :] = torch.randn(20, 1) * 2.0
y_known = torch.sin(X @ W_known)

# 未知の宇宙的危機（タスク2）：残りの80次元（未知の科学・多様性）に依存
W_unknown = torch.zeros(input_dim, 1)
W_unknown[20:, :] = torch.randn(80, 1) * 2.0
y_unknown = torch.relu(X @ W_unknown)

X_train, X_crisis = X[:1500], X[1500:]
y_known_train, _ = y_known[:1500], y_known[1500:]
y_unknown_crisis = y_unknown[1500:]

# ==========================================
# 2. α（贅沢品への投資割合）の動的スケジュール
# ==========================================
epochs_peacetime = 300
epochs_crisis = 80
total_epochs = epochs_peacetime + epochs_crisis

alphas = np.zeros(total_epochs)

# 平時：ベースライン0.5を中心に、ランダムウォーク的に揺らぐ（探索状態）
alpha_val = 0.5
for i in range(epochs_peacetime):
    alpha_val += np.random.normal(0, 0.05)
    alpha_val = np.clip(alpha_val, 0.2, 0.8) # 20%〜80%の間を揺らぐ
    alphas[i] = alpha_val

# 危機時：エントロピーの減少に伴い、急速に 0（必需品・生存）へ収束
for i in range(epochs_peacetime, total_epochs):
    alpha_val *= 0.8 # 指数関数的な減衰（搾取状態への移行）
    alphas[i] = alpha_val

# ==========================================
# 3. 社会モデルの定義
# ==========================================
class SocialBackbone(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64), nn.ReLU(),
            nn.Linear(64, hidden_dim), nn.ReLU()
        )
    def forward(self, x): return self.net(x)

# 社会C：ハイブリッドモデル（必需品と贅沢品の両方を α の割合で学習）
class SocietyHybrid(nn.Module):
    def __init__(self, backbone):
        super().__init__()
        self.backbone = backbone
        self.head_known = nn.Linear(hidden_dim, 1) # 必需品（既知タスク）用
        self.decoder = nn.Sequential(              # 贅沢品（多様性の探索）用
            nn.Linear(hidden_dim, 64), nn.ReLU(),
            nn.Linear(64, input_dim)
        )
    def forward_known(self, x): return self.head_known(self.backbone(x))
    def forward_recon(self, x): return self.decoder(self.backbone(x))

class CrisisSurvival(nn.Module):
    def __init__(self, trained_backbone):
        super().__init__()
        self.backbone = trained_backbone
        self.survival_head = nn.Linear(hidden_dim, 1)
    def forward(self, x):
        with torch.no_grad(): # 危機時は潜在空間を組み替える猶予がない
            features = self.backbone(x)
        return self.survival_head(features)

# ==========================================
# 4. フェーズ1：平時（数十年の歴史）
# ==========================================
print("--- フェーズ1：平時（αの揺らぎによる学習） ---")
# 各社会のバックボーン
bb_A, bb_B, bb_C = SocialBackbone(), SocialBackbone(), SocialBackbone()

# 社会A（α=0固定: 既知タスクのみ）、社会B（α=1固定: 探索のみ）
head_A = nn.Linear(hidden_dim, 1)
decoder_B = nn.Sequential(nn.Linear(hidden_dim, 64), nn.ReLU(), nn.Linear(64, input_dim))
model_C = SocietyHybrid(bb_C)

opt_A = optim.Adam(list(bb_A.parameters()) + list(head_A.parameters()), lr=0.01, weight_decay=1e-4)
opt_B = optim.Adam(list(bb_B.parameters()) + list(decoder_B.parameters()), lr=0.01, weight_decay=1e-4)
opt_C = optim.Adam(model_C.parameters(), lr=0.01, weight_decay=1e-4)

criterion = nn.MSELoss()

for epoch in range(epochs_peacetime):
    # 社会A：必需品に全振り
    loss_A = criterion(head_A(bb_A(X_train)), y_known_train)
    opt_A.zero_grad(); loss_A.backward(); opt_A.step()

    # 社会B：贅沢品に全振り
    loss_B = criterion(decoder_B(bb_B(X_train)), X_train)
    opt_B.zero_grad(); loss_B.backward(); opt_B.step()

    # 社会C：揺らぐ α_t に基づくハイブリッド学習
    alpha_t = alphas[epoch]
    loss_C_known = criterion(model_C.forward_known(X_train), y_known_train)
    loss_C_recon = criterion(model_C.forward_recon(X_train), X_train)
    # 損失関数の動的ブレンド
    loss_C = (1 - alpha_t) * loss_C_known + alpha_t * loss_C_recon

    opt_C.zero_grad(); loss_C.backward(); opt_C.step()

print(f"平時終了時のLoss - 社会A: {loss_A:.4f}, 社会B: {loss_B:.4f}, 社会C: {loss_C:.4f}")

# ==========================================
# 5. フェーズ2：未知の危機発生（一斉に生存タスクへ）
# ==========================================
print("--- フェーズ2：未知の危機発生！（α→0への収束） ---")
surv_A, surv_B, surv_C = CrisisSurvival(bb_A), CrisisSurvival(bb_B), CrisisSurvival(bb_C)

opt_surv_A = optim.Adam(surv_A.survival_head.parameters(), lr=0.02)
opt_surv_B = optim.Adam(surv_B.survival_head.parameters(), lr=0.02)
opt_surv_C = optim.Adam(surv_C.survival_head.parameters(), lr=0.02)

hist_A, hist_B, hist_C = [], [], []

for epoch in range(epochs_crisis):
    # 危機時は全社会が「未知の危機」のロスを下げることに全振りする（α実質0）
    l_A = criterion(surv_A(X_crisis), y_unknown_crisis)
    opt_surv_A.zero_grad(); l_A.backward(); opt_surv_A.step(); hist_A.append(l_A.item())

    l_B = criterion(surv_B(X_crisis), y_unknown_crisis)
    opt_surv_B.zero_grad(); l_B.backward(); opt_surv_B.step(); hist_B.append(l_B.item())

    l_C = criterion(surv_C(X_crisis), y_unknown_crisis)
    opt_surv_C.zero_grad(); l_C.backward(); opt_surv_C.step(); hist_C.append(l_C.item())

print(f"危機対応終了時のLoss - 社会A: {l_A:.4f}, 社会B: {l_B:.4f}, 社会C: {l_C:.4f}")

# ==========================================
# 6. 結果の可視化
# ==========================================
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# グラフ1: αの軌跡
ax1.plot(range(total_epochs), alphas, color='purple', linewidth=2)
ax1.axvline(x=epochs_peacetime, color='red', linestyle='--', label='Unknown Crisis Occurs')
ax1.set_title('Dynamics of \u03B1 (Investment Ratio in Luxuries / Exploration)')
ax1.set_xlabel('Time (Epochs)')
ax1.set_ylabel('\u03B1 (0: Pure Necessity, 1: Pure Luxury)')
ax1.fill_between(range(epochs_peacetime), alphas[:epochs_peacetime], color='purple', alpha=0.1, label='Peacetime: Random Fluctuation')
ax1.fill_between(range(epochs_peacetime, total_epochs), alphas[epochs_peacetime:], color='red', alpha=0.1, label='Crisis: Collapse to 0 (Exploitation)')
ax1.legend()
ax1.grid(True)

# グラフ2: 危機対応時のLoss比較
ax2.plot(hist_A, label='Society A (\u03B1=0: Totalitarian / Over-fit)', color='red', linestyle='--', linewidth=2)
ax2.plot(hist_B, label='Society B (\u03B1=1: Pure Diversity)', color='blue', linestyle=':', linewidth=2)
ax2.plot(hist_C, label='Society C (Dynamic \u03B1: Hybrid / Adaptive)', color='green', linewidth=2)
ax2.set_title('Survival in Unknown Crisis: Comparison of Societal Models')
ax2.set_xlabel('Crisis Response Time (Epochs / SGD steps)')
ax2.set_ylabel('Crisis Damage (MSE Loss - lower is better)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
````

* **社会A（全体主義）：贅沢品への投資 $\alpha = 0$**
    平時の学習フェーズにおいて、社会Aは `y_known`（既知のタスク＝必需品）に対する Loss だけを最小化しています。つまり、100%のリソースを必需品に固定して投資しています。

具体的には、損失関数（社会の目的）を以下のように足し合わせます。

$$Loss_{total} = (1 - \alpha) \cdot Loss_{Necessity} + \alpha \cdot Loss_{Luxury}$$

* **平時：** 未知の危機（真の損失）が見えないため、$\alpha$ は高い値でランダムに揺らぎます（贅沢品による探索）。
* **危機の近傍：** 環境からのシグナル（エントロピーの減少）を検知すると、社会の生存本能（メタ学習のメカニズム）が働き、$\alpha \to 0$ に向けて急速にパラメーターを絞り込みます（探索から利用＝必需品への転化）。

コードにおける `X` と `y` は、AI（機械学習）の枠組みを社会の生存モデルに当てはめると、それぞれ以下のような意味を持ちます。

### 1. `X` の正体：「世界の複雑さ（環境からの入力シグナル）」
`X`（100次元のベクトル）は、社会が直面している**「世界（環境）から絶えず入力される無数の情報・資源・現象のシグナル」**を表しています。

### 2. `y` の正体：「生存に必要な正解（出力すべき行動や必需品の量）」
一方、`y` はその環境 `X` の下で、危機を乗り越えて生き残るために社会が叩き出さなければならない**「正解の行動（あるいは必要な特定の必需品の量）」**を表しています。

ネットワークが予測した値（出力）を $\hat{y}$ としたとき、$\hat{y}$ が本当の $y$ と一致していれば Loss が $0$（生存）、外れていれば Loss が増大（絶滅の危機）となります。

* **`y_known`（既知の危機の正解）**: 過去の経験からわかっている「こうすれば生き残れる」という正解です。コード内で `W_known[:20, :]` としたように、これは世界の100次元の情報のうち、**「特定の20次元（例えば重工業力、食料生産力、軍事力など）」**だけを見れば計算できる正解です。
    * 社会A（全体主義）は、ひたすら `X` からこの `y_known` を出力する訓練だけを行います（過剰適応）。その結果、残りの80次元の情報を「生存（既知のタスク）に無関係な無駄」としてネットワークの重みから削ぎ落としてしまいます。
* **`y_unknown`（未知の宇宙的危機の正解）**: 突然やってきた新しい危機を乗り越えるための正解です。これは `W_unknown[20:, :]` としたように、社会Aが「無駄だ」と切り捨てていた**「残りの80次元（例えば、高度なシミュレーション技術、基礎生物学、柔軟な思想など）」**の情報を使って計算しなければならない正解です。

### 全体の構造まとめ

あなたのモデルにおける「必需品と贅沢品」の関係を、機械学習の言葉で整理するとこうなります。

* **危機の到来**: 新しい生存のルール $y_{unknown}$ が突きつけられること。このとき、社会Aは必要な次元をすでに忘却しており計算不能（即死）になりますが、社会Bは $X$ 全体の概念を「潜在空間」に保持していたため、少しの調整（SGD）で正解 $y_{unknown}$ を弾き出すことができるのです。

````python
"""
必需品と贅沢品の宇宙的独立関係：拡張版
========================================
Gemini版からの主な変更点：
1. エントロピーの明示的計算（αの分布から毎ステップ測定）
2. ξ₁（サンプリングノイズ）とξ₂（構造的不確実性）の分離
3. 危機フェーズをPolicy Gradient的な強化学習に変更
4. 4パネル可視化（α軌跡・エントロピー・loss比較・ξ₁ξ₂）

Google Colab で実行可能。
必要ライブラリ：torch, numpy, matplotlib（Colab標準搭載）
"""

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
from collections import deque

torch.manual_seed(42)
np.random.seed(42)

# ══════════════════════════════════════════
# 0. ヘルパー：エントロピー計算
# ══════════════════════════════════════════
def entropy_from_alpha(alpha_history, window=30):
    """
    αの履歴からエントロピーを推定する。
    αをビン分割して経験的分布を作り Shannon entropy を計算。
    これが ξ₂（構造的不確実性）の代理変数になる。
    """
    if len(alpha_history) < window:
        return np.log(10)  # 最大不確実性
    recent = alpha_history[-window:]
    counts, _ = np.histogram(recent, bins=10, range=(0, 1))
    probs = counts / counts.sum()
    probs = probs[probs > 0]
    return -np.sum(probs * np.log(probs))

def xi2_from_crisis_distance(d, d_max=1.0):
    """
    ξ₂：危機への距離 d に基づく構造的不確実性。
    d→0 で ξ₂→0（ナイト的不確実性の解消）。
    """
    return (d / d_max) ** 0.7

# ══════════════════════════════════════════
# 1. 環境の定義（Gemini版を継承・拡張）
# ══════════════════════════════════════════
input_dim  = 100
hidden_dim = 20
num_samples = 2000

X = torch.randn(num_samples, input_dim)

# 既知タスク（必需品）：最初の20次元
W_known = torch.zeros(input_dim, 1)
W_known[:20, :] = torch.randn(20, 1) * 2.0
y_known = torch.sin(X @ W_known)

# 未知の危機タスク（残り80次元）
W_unknown = torch.zeros(input_dim, 1)
W_unknown[20:, :] = torch.randn(80, 1) * 2.0
y_unknown = torch.relu(X @ W_unknown)

X_train, X_crisis = X[:1500], X[1500:]
y_known_train     = y_known[:1500]
y_unknown_crisis  = y_unknown[1500:]

# ══════════════════════════════════════════
# 2. α のダイナミクス（エントロピー連動）
# ══════════════════════════════════════════
epochs_peacetime = 300
epochs_crisis    = 80
total_epochs     = epochs_peacetime + epochs_crisis
d_max            = 1.0

alphas    = np.zeros(total_epochs)
xi1_arr   = np.zeros(total_epochs)   # サンプリングノイズ（学習で減少）
xi2_arr   = np.zeros(total_epochs)   # 構造的不確実性（危機で消滅）
H_arr     = np.zeros(total_epochs)   # エントロピー推定値

alpha_val   = 0.5
alpha_hist  = []
# ξ₁ は学習エポックとともに指数減衰（事前学習による収束）
xi1_base    = 0.08

for i in range(epochs_peacetime):
    # 危機への距離：平時は遠い（1.0）→ 危機直前で急速に縮まる
    d = d_max * (1 - (i / epochs_peacetime) ** 3)

    # ξ₁：学習進行で減少
    xi1 = xi1_base * np.exp(-i / 150) + 0.01
    # ξ₂：危機への距離で決まる
    xi2 = xi2_from_crisis_distance(d, d_max)

    # α更新：ξ₁のノイズ + ξ₂による揺らぎ幅
    noise = np.random.normal(0, xi1 + xi2 * 0.15)
    alpha_val = np.clip(alpha_val + noise, 0.1, 0.9)
    alpha_hist.append(alpha_val)

    alphas[i]  = alpha_val
    xi1_arr[i] = xi1
    xi2_arr[i] = xi2
    H_arr[i]   = entropy_from_alpha(alpha_hist)

# 危機フェーズ：ξ₂→0 に伴い α→α*（危機が明かす正解）へ収束
alpha_star = 0.1   # 危機時の最適配分（必需品寄り）
for i in range(epochs_peacetime, total_epochs):
    j = i - epochs_peacetime
    d = d_max * max(0, 1 - j / 20) ** 2   # 急速に0へ

    xi1 = xi1_base * np.exp(-epochs_peacetime / 150) * 0.5
    xi2 = xi2_from_crisis_distance(d, d_max)

    # 危機：報酬信号が α を α* へ引き寄せる（PG的更新）
    reward_signal = -(alpha_val - alpha_star)   # 正解方向への勾配
    alpha_val = np.clip(
        alpha_val + 0.15 * reward_signal + np.random.normal(0, xi1),
        0.0, 1.0
    )
    alpha_hist.append(alpha_val)

    alphas[i]  = alpha_val
    xi1_arr[i] = xi1
    xi2_arr[i] = xi2
    H_arr[i]   = entropy_from_alpha(alpha_hist)

# ══════════════════════════════════════════
# 3. 社会モデル（Gemini版を継承）
# ══════════════════════════════════════════
class SocialBackbone(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64), nn.ReLU(),
            nn.Linear(64, hidden_dim), nn.ReLU()
        )
    def forward(self, x):
        return self.net(x)

class SocietyHybrid(nn.Module):
    """社会C：α に応じて必需品と贅沢品のlossをブレンド"""
    def __init__(self, backbone):
        super().__init__()
        self.backbone = backbone
        self.head_known  = nn.Linear(hidden_dim, 1)
        self.decoder     = nn.Sequential(
            nn.Linear(hidden_dim, 64), nn.ReLU(),
            nn.Linear(64, input_dim)
        )
    def forward_known(self, x):
        return self.head_known(self.backbone(x))
    def forward_recon(self, x):
        return self.decoder(self.backbone(x))

class CrisisSurvival(nn.Module):
    """危機対応：事前学習済みbackboneを凍結し、survival headのみ更新"""
    def __init__(self, trained_backbone):
        super().__init__()
        self.backbone      = trained_backbone
        self.survival_head = nn.Linear(hidden_dim, 1)
    def forward(self, x):
        with torch.no_grad():
            features = self.backbone(x)
        return self.survival_head(features)

# ══════════════════════════════════════════
# 4. フェーズ1：平時学習
# ══════════════════════════════════════════
print("--- フェーズ1：平時（事前学習＝析出フェーズ） ---")
bb_A, bb_B, bb_C = SocialBackbone(), SocialBackbone(), SocialBackbone()
head_A    = nn.Linear(hidden_dim, 1)
decoder_B = nn.Sequential(
    nn.Linear(hidden_dim, 64), nn.ReLU(), nn.Linear(64, input_dim)
)
model_C   = SocietyHybrid(bb_C)

opt_A = optim.Adam(list(bb_A.parameters()) + list(head_A.parameters()),
                   lr=0.01, weight_decay=1e-4)
opt_B = optim.Adam(list(bb_B.parameters()) + list(decoder_B.parameters()),
                   lr=0.01, weight_decay=1e-4)
opt_C = optim.Adam(model_C.parameters(),
                   lr=0.01, weight_decay=1e-4)

criterion = nn.MSELoss()

# loss_C の履歴（α の寄与を分解して記録）
loss_C_known_hist = []
loss_C_recon_hist = []
loss_C_total_hist = []

for epoch in range(epochs_peacetime):
    loss_A = criterion(head_A(bb_A(X_train)), y_known_train)
    opt_A.zero_grad(); loss_A.backward(); opt_A.step()

    loss_B = criterion(decoder_B(bb_B(X_train)), X_train)
    opt_B.zero_grad(); loss_B.backward(); opt_B.step()

    alpha_t     = alphas[epoch]
    lc_known    = criterion(model_C.forward_known(X_train), y_known_train)
    lc_recon    = criterion(model_C.forward_recon(X_train), X_train)
    # ──────────────────────────────────────────────────────────
    # loss_C = (1-α)・必需品loss + α・贅沢品loss
    # α が揺れることで、バックボーンは「どちらにも使える表現」を学ぶ。
    # これが事前学習における「析出」に相当する。
    # ──────────────────────────────────────────────────────────
    loss_C = (1 - alpha_t) * lc_known + alpha_t * lc_recon
    opt_C.zero_grad(); loss_C.backward(); opt_C.step()

    loss_C_known_hist.append(lc_known.item())
    loss_C_recon_hist.append(lc_recon.item())
    loss_C_total_hist.append(loss_C.item())

print(f"  社会A loss: {loss_A.item():.4f} | 社会B loss: {loss_B.item():.4f} | 社会C loss: {loss_C.item():.4f}")

# ══════════════════════════════════════════
# 5. フェーズ2：危機（強化学習フェーズ）
# ══════════════════════════════════════════
print("--- フェーズ2：未知の危機発生（強化学習フェーズ） ---")
surv_A = CrisisSurvival(bb_A)
surv_B = CrisisSurvival(bb_B)
surv_C = CrisisSurvival(bb_C)

# 危機フェーズ：survival headのみ学習率を高めに設定
opt_sA = optim.Adam(surv_A.survival_head.parameters(), lr=0.03)
opt_sB = optim.Adam(surv_B.survival_head.parameters(), lr=0.03)
opt_sC = optim.Adam(surv_C.survival_head.parameters(), lr=0.03)

hist_A, hist_B, hist_C = [], [], []

# Policy Gradient 風の報酬スケーリング
# 報酬 = -loss（lossが下がるほど報酬大）
# ベースライン：移動平均でバリアンス削減
baseline_window = 10
baselines_A = deque(maxlen=baseline_window)
baselines_B = deque(maxlen=baseline_window)
baselines_C = deque(maxlen=baseline_window)

for epoch in range(epochs_crisis):
    # ─── 社会A ───
    pred_A  = surv_A(X_crisis)
    loss_A  = criterion(pred_A, y_unknown_crisis)
    bl_A    = np.mean(baselines_A) if baselines_A else loss_A.item()
    # 報酬がベースライン以上なら通常更新、以下なら強調更新
    reward_scale_A = 1.0 + max(0, bl_A - loss_A.item())
    opt_sA.zero_grad()
    (loss_A * reward_scale_A).backward()
    opt_sA.step()
    baselines_A.append(loss_A.item())
    hist_A.append(loss_A.item())

    # ─── 社会B ───
    pred_B  = surv_B(X_crisis)
    loss_B  = criterion(pred_B, y_unknown_crisis)
    bl_B    = np.mean(baselines_B) if baselines_B else loss_B.item()
    reward_scale_B = 1.0 + max(0, bl_B - loss_B.item())
    opt_sB.zero_grad()
    (loss_B * reward_scale_B).backward()
    opt_sB.step()
    baselines_B.append(loss_B.item())
    hist_B.append(loss_B.item())

    # ─── 社会C ───
    pred_C  = surv_C(X_crisis)
    loss_C  = criterion(pred_C, y_unknown_crisis)
    bl_C    = np.mean(baselines_C) if baselines_C else loss_C.item()
    reward_scale_C = 1.0 + max(0, bl_C - loss_C.item())
    opt_sC.zero_grad()
    (loss_C * reward_scale_C).backward()
    opt_sC.step()
    baselines_C.append(loss_C.item())
    hist_C.append(loss_C.item())

print(f"  社会A loss: {hist_A[-1]:.4f} | 社会B loss: {hist_B[-1]:.4f} | 社会C loss: {hist_C[-1]:.4f}")

# ══════════════════════════════════════════
# 6. 可視化（4パネル）
# ══════════════════════════════════════════
fig = plt.figure(figsize=(14, 12))
fig.patch.set_facecolor('#0d0d1a')
gs  = gridspec.GridSpec(2, 2, figure=fig,
                        hspace=0.42, wspace=0.35,
                        left=0.08, right=0.97,
                        top=0.92, bottom=0.07)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

DARK_BG  = '#0d0d1a'
C_PEACE  = '#4ecdc4'
C_CRISIS = '#ff6b6b'
C_A      = '#ff6b6b'
C_B      = '#4ecdc4'
C_C      = '#ffe66d'
C_ENT    = '#a78bfa'
C_XI1    = '#67e8f9'
C_XI2    = '#f97316'

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_facecolor(DARK_BG)
    for sp in ax.spines.values():
        sp.set_color('#334')
    ax.tick_params(colors='#aab', labelsize=9)
    ax.xaxis.label.set_color('#ccd')
    ax.yaxis.label.set_color('#ccd')
    ax.title.set_color('#eef')
    ax.grid(True, color='#1e2040', lw=0.6)

fig.suptitle('必需品と贅沢品の宇宙的独立関係：拡張版\n'
             '（エントロピー・二層ノイズ・強化学習フェーズ）',
             fontsize=13, color='#eef8ff', fontweight='bold', y=0.98)

ts = np.arange(total_epochs)

# ── Panel 1：α の軌跡 ──
ax1.fill_between(range(epochs_peacetime),
                 alphas[:epochs_peacetime],
                 color=C_PEACE, alpha=0.15,
                 label='平時：ξ1+ξ2 によるランダム揺動')
ax1.fill_between(range(epochs_peacetime, total_epochs),
                 alphas[epochs_peacetime:],
                 color=C_CRISIS, alpha=0.15,
                 label='危機：報酬信号で α* へ収束')
ax1.plot(ts, alphas, color=C_PEACE, lw=1.5)
ax1.plot(range(epochs_peacetime, total_epochs),
         alphas[epochs_peacetime:], color=C_CRISIS, lw=2.0)
ax1.axvline(epochs_peacetime, color=C_CRISIS, lw=1.5, ls='--', alpha=0.8)
ax1.axhline(alpha_star, color=C_C, lw=1.0, ls=':',
            label=f'α* = {alpha_star}（危機が明かす正解）')
ax1.text(epochs_peacetime + 2, 0.75, '危機発生', color=C_CRISIS, fontsize=8)
ax1.set_title('① α の軌跡\n平時揺動 → 危機収束（PG報酬信号）')
ax1.set_xlabel('時間（エポック）')
ax1.set_ylabel('α（0:必需品 ／ 1:贅沢品）')
ax1.set_ylim(0, 1)
ax1.legend(fontsize=7.5, facecolor='#1a1a2e',
           labelcolor='#ccd', framealpha=0.7, loc='upper right')

# ── Panel 2：エントロピーと ξ₁・ξ₂ ──
ax2_twin = ax2.twinx()
ax2_twin.set_facecolor(DARK_BG)
ax2_twin.tick_params(colors='#aab', labelsize=9)
ax2_twin.yaxis.label.set_color('#ccd')

ax2.plot(ts, H_arr,   color=C_ENT, lw=2.0, label='H（α のエントロピー）')
ax2.fill_between(ts, 0, H_arr, color=C_ENT, alpha=0.12)
ax2_twin.plot(ts, xi1_arr, color=C_XI1, lw=1.5, ls='--',
              label='ξ1（サンプリングノイズ）', alpha=0.85)
ax2_twin.plot(ts, xi2_arr, color=C_XI2, lw=1.5, ls=':',
              label='ξ2（構造的不確実性）', alpha=0.85)
ax2.axvline(epochs_peacetime, color=C_CRISIS, lw=1.5, ls='--', alpha=0.8)

# 二つの凡例を合成
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2,
           fontsize=7.5, facecolor='#1a1a2e',
           labelcolor='#ccd', framealpha=0.7, loc='upper right')

ax2.set_title('② エントロピー H と ξ1・ξ2\nξ1は学習で減少、ξ2は危機で消滅')
ax2.set_xlabel('時間（エポック）')
ax2.set_ylabel('H（エントロピー）', color=C_ENT)
ax2_twin.set_ylabel('ノイズ強度 ξ', color='#aab')
ax2_twin.spines['right'].set_color('#334')

# ── Panel 3：loss_C の分解（α の意味を可視化）──
ep3 = np.arange(epochs_peacetime)
ax3.plot(ep3, loss_C_known_hist, color=C_A, lw=1.5, alpha=0.8,
         label='必需品 loss（head_known）')
ax3.plot(ep3, loss_C_recon_hist, color=C_B, lw=1.5, alpha=0.8,
         label='贅沢品 loss（decoder_recon）')
ax3.plot(ep3, loss_C_total_hist, color=C_C, lw=2.0,
         label='loss_C = (1-α)·必需品 + α·贅沢品')

# α の値を背景色で表現
for i in range(0, epochs_peacetime - 1, 3):
    a = alphas[i]
    ax3.axvspan(i, i+3,
                color=C_PEACE if a > 0.5 else C_CRISIS,
                alpha=a * 0.04)

ax3.set_title('③ loss_C の分解（社会Cの内部構造）\n'
              'α揺動 → backbone が「両用表現」を学ぶ')
ax3.set_xlabel('平時エポック')
ax3.set_ylabel('Loss（MSE）')
ax3.legend(fontsize=7.5, facecolor='#1a1a2e',
           labelcolor='#ccd', framealpha=0.7)

# ── Panel 4：危機対応 loss 比較 ──
ep4 = np.arange(epochs_crisis)
ax4.plot(ep4, hist_A, color=C_A, lw=2.0, ls='--',
         label='社会A（α=0固定：必需品全振り）')
ax4.plot(ep4, hist_B, color=C_B, lw=2.0, ls=':',
         label='社会B（α=1固定：贅沢品全振り）')
ax4.plot(ep4, hist_C, color=C_C, lw=2.5,
         label='社会C（動的α：ハイブリッド）')

# 最終loss の注釈
for hist, col, name in [(hist_A, C_A, 'A'),
                         (hist_B, C_B, 'B'),
                         (hist_C, C_C, 'C')]:
    ax4.annotate(f'{hist[-1]:.3f}',
                 xy=(epochs_crisis - 1, hist[-1]),
                 xytext=(epochs_crisis - 15, hist[-1] + 0.02),
                 color=col, fontsize=8,
                 arrowprops=dict(arrowstyle='->', color=col, lw=0.8))

ax4.set_title('④ 危機対応 loss（強化学習フェーズ）\n'
              '社会Cの事前学習backbone が奏功する')
ax4.set_xlabel('危機対応時間（エポック）')
ax4.set_ylabel('危機ダメージ（MSE Loss）')
ax4.legend(fontsize=7.5, facecolor='#1a1a2e',
           labelcolor='#ccd', framealpha=0.7)

plt.savefig('basics_luxuries_v2.png',
            dpi=150, bbox_inches='tight',
            facecolor=DARK_BG)
plt.show()
print("\n保存完了：basics_luxuries_v2.png")
print("\n── 最終まとめ ──")
print(f"  危機対応 final loss  社会A: {hist_A[-1]:.4f}  社会B: {hist_B[-1]:.4f}  社会C: {hist_C[-1]:.4f}")
print(f"  社会Cの優位性: {min(hist_A[-1], hist_B[-1]) / hist_C[-1]:.2f}x（C比）")
````

これでは、エントロピーはただの「結果の影」であり、「そこから学習する」という本質を体現していません。

このズレをどう解釈すべきか、そして真に「エントロピーから学習する」とはどういうことなのか、AIの最先端の理論（**最大エントロピー強化学習**と**情報量最大化**）を用いて解きほぐします。

---

### 1. 現在のコードにおけるエントロピーの扱い（単なる温度計）
現在のコードで計算されているエントロピー $H$ は、「社会がどれだけ $\alpha$（配分比率）をフラフラと揺らしているか」の指標です。
* **平時：** $\xi_1$ と $\xi_2$ という外部ノイズによって $\alpha$ が揺らされているため、「結果として」エントロピー $H$ が高くなっています。
* **危機時：** 報酬信号によって強制的に $\alpha \to 0$ に引き寄せられるため、「結果として」エントロピー $H$ がゼロに潰れます。

ここでは、エントロピーは学習を駆動しておらず、単に「社会の迷い具合」を外から測定しているだけです。

### 2. 真の姿：「最大エントロピー強化学習（MaxEnt RL）」
もし、私たちの哲学をプログラムの「学習の駆動力（Loss関数）」として正しく実装するなら、最先端の強化学習アルゴリズムである「Soft Actor-Critic（SAC）」などに使われる**最大エントロピー原理**を導入する必要があります。

本来、社会Cの目的関数（生存戦略）は、以下のようになっているべきです。

$$\max_{\pi} \mathbb{E} \left[ \text{生存報酬} + \lambda \cdot \mathbf{H}(\pi) \right]$$

* $\pi$ は社会の行動（$\alpha$ の決定方策）です。
* $\lambda$ は温度パラメーター（不確実性の大きさ）です。

この数式の意味は、**「目先の生存報酬を最大化しつつ、同時に『選択肢の多様性（エントロピー $H$）』も意図的に最大化せよ」**という命令です。
つまり、エントロピーが高いこと（＝何にでもなれる状態を維持すること）自体が、システムにとっての**「報酬（目的）」**として扱われます。平時において $\lambda$（未知への警戒感）が大きい場合、システムはあえて特定の $\alpha$ に固定されることを嫌がり、自ら進んで $\alpha$ を揺らがせ、贅沢品への投資を維持しようとします。

これが「エントロピーから学習する（エントロピーを高く保つように学習する）」の本来の姿です。

### 3. 「析出」とエントロピー：相互情報量の最大化
