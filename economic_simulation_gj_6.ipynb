メインのシミュレーション・プログラムの最終稿である。説明は追ってするので、ここでは説明しない。しかし、読んでいただけるなら、かなり動きはわかるはずである。

````python
#esgj06
import numpy as np
import matplotlib.pyplot as plt

class EconomicAgent:
    def __init__(self, n_people, initial_asset, label):
        self.n_people = n_people
        self.asset_stock = initial_asset # 株式
        self.asset_bond = 0.0           # 債券
        self.income = 0
        self.label = label
        self.happiness = 50.0 # 初期幸福度

    def consume(self, income, price_basic, price_luxury, variety, tech_threat, avg_asset_total, income_ratio=1.0):
        """
        消費行動と幸福度の算出
        variety: 文化が生み出す選択肢の多さ
        tech_threat: 技術偏重による脅威度
        income_ratio: 大衆の相対的な所得状況
        """
        # 1. 必需品の確保 (生存目標)
        base_req_unit = 10.0
        # 格差が是正される（ratioが上がる）ほど、大衆は必需品の「質」や「確実な確保」を求め、需要が強気化する。
        # 一方で、あまりに格差が大きい（ratioが低い）と、生きるための最低限しか求められない。
        if self.label == "Mass":
            # 所得比率が改善するほど、必需品への要求水準が動的に上昇する（潜在需要の顕在化）
            dynamic_req_unit = base_req_unit * (1.0 + 0.8 * np.tanh(income_ratio * 2.0))
        else:
            # 富裕層はすでに飽和している
            dynamic_req_unit = base_req_unit * 1.2

        total_req_basic = dynamic_req_unit * self.n_people
        cost_basic = total_req_basic * price_basic

        real_basic = 0
        budget_luxury = 0

        if income >= cost_basic:
            real_basic = total_req_basic
            budget_luxury = income - cost_basic
        else:
            real_basic = (income / (price_basic + 1e-9))
            budget_luxury = 0

        # 2. 贅沢品の購入
        # ここに「選択の余地」が効いてくる。varietyが多いほど効用が高い（＝株式価値の源泉）
        real_luxury = 0
        if price_luxury > 0:
            real_luxury = budget_luxury / price_luxury

        # 3. 幸福度の計算（制約条件）
        # 基礎充足
        satisfaction_basic = real_basic / (total_req_basic + 1e-9)

        # 選択の自由: 多様性が「富」として認識される本質（文化レベル * 消費量）
        # ただし、選択肢があっても買えなければ意味がない
        # JRF説: 「選択の余地があることが幸福」
        utility_choice = np.log1p(real_luxury * (variety + 1e-9) / 1000.0)

        # 相対的格差の影響 (ペナルティ)
        # 自分の資産が平均より著しく低い場合の幸福度低下
        current_total_asset = (self.asset_stock + self.asset_bond) / self.n_people
        inequality_penalty = max(0, (avg_asset_total - current_total_asset) / (avg_asset_total + 1e-9)) * 5

        # 幸福度 = 充足 + 自由 - 脅威 - 格差
        # （文化なき技術は不幸）
        target_happiness = (35 * satisfaction_basic) + (45 * utility_choice) - tech_threat - inequality_penalty
        self.happiness = self.happiness * 0.8 + target_happiness * 0.2

        return {
            'real_basic': real_basic,
            'real_luxury': real_luxury,
            'luxury_budget': budget_luxury,
            'basic_demand': total_req_basic,
        }

class ComplexEconomySimulation:
    def __init__(self, rate_scenario="low_stable", tech_bias=0.0, public_work_rate=0.0, redistribution_rate=0.2):
        self.steps = 100
        self.tech_level = 2.0     # 技術ストック（生産性）
        self.culture_level = 2.0  # 文化ストック（多様性・選択肢）
        self.stock_market_val = 1500.0 # 株式時価総額
        self.bond_market_val = 1500.0    # 債券市場の初期値

        self.tech_bias = tech_bias # 投資の技術偏重度
        self.redistribution_rate = redistribution_rate
        self.rate_scenario = rate_scenario
        self.public_work_rate = public_work_rate

        self.labor_total = 1000.0
        self.price_basic = 1.0
        self.price_luxury = 5.0
        self.bank_premium = 0.02 # 初期銀行プレミアム

        # エージェント：富裕層（株主）と大衆層（労働者）
        # 富裕層は株式を保有し、その評価益で豊かになる
        self.rich = EconomicAgent(10, 8000.0, "Rich")
        self.mass = EconomicAgent(90, 200.0, "Mass")

        self.history = {
            'stock_val': [],
            'bond_val': [],
            'happiness_mass': [],
            'happiness_rich': [],
            'price_basic': [],
            'price_luxury': [],
            'demand_basic': [],
            'relative_price': [],
            'supply_basic': [],
            'supply_luxury': [],
            'labor_share_basic': [],
            'tech_level': [],
            'public_demand_share': [],
            'asset_ratio_stock': [],
            'policy_rate': [],
            'bank_premium': [],
            'culture_level': [],
            'luxury_rate': [],
            'inequality_index': [],
        }

    def run(self):
        for t in range(self.steps):
            # 1. 金利の設定
            if self.rate_scenario == "high_rate_shock":
                # 途中から金利が急上昇するシナリオ
                policy_rate = 0.04 if t < 40 else 0.15
                bank_premium_base = 0.08
            elif self.rate_scenario == "bank_premium_shock":
                policy_rate = 0.04
                bank_premium_base =  0.08 if t < 40 else 0.15
            elif self.rate_scenario == "low_stable":
                policy_rate = 0.04
                bank_premium_base = 0.08
            else:
                policy_rate = 0.05
                bank_premium_base = 0.08

            # 2. 銀行プレミアム (JRFモデル: 文化不足で上昇)
            # 選択の余地（文化）が少ないほど、銀行プレミアム（摩擦コスト）が増大する
            # また、技術革新のスピード（tech_levelの増加分）が大きいほど一時的な赤字が増える
            friction_reduction = np.log1p(self.culture_level)
            innovation_friction = np.log1p(self.tech_level)
            # 公共事業(強制的な必需品支出)があるほど、プレミアム上昇が抑えられるロジック
            public_stability = 1.0 + self.public_work_rate * 2.0
            self.bank_premium = (bank_premium_base * (innovation_friction / (friction_reduction + 0.1))) / public_stability


            effective_rate = policy_rate + self.bank_premium

            # 3. 公共事業
            # 公共事業は国債(債券)で賄われる
            public_demand = 1000.0 * self.public_work_rate
            # 債券発行残高の増大（これが市場のパイを広げる）
            self.bond_market_val += public_demand * 0.05

            # 4. 在庫・ストックの更新 (金利によるスパース化)
            # 金利が高いと、在庫（将来価値）の保持が難しくなる (スパース化)
            discount_factor = 1.0 / (1.0 + effective_rate)
            # 株式投資は金利に敏感、債券は安定
            # ストックが高いほど成長は鈍化する（収穫逓減）
            invest_fund = np.sqrt(self.stock_market_val) * 3.0 * discount_factor

            # 技術への投資（バイアスに基づく）
            self.tech_level += (invest_fund * self.tech_bias) * 0.06 / (1 + self.tech_level * 0.005)

            # 文化への投資
            # 技術の発展の否定＝文化の発展とならないよう
            # 文化への投資から公共事業分も抜いておく
            self.culture_level += (invest_fund * max(0, 1 - self.tech_bias - self.public_work_rate) * 0.06) / (1 + self.culture_level * 0.005)

            # 5. 生産と労働配分（構造的変化の肝）
            # 生産性の向上をマイルドに。
            # 文化レベルが技術レベルを上げる側面があると思われるが、
            # 同時に文化レベルが質の向上を求める面もあると思われる。
            # 文化レベルの生産性への寄与は質の向上を求めるほうが
            # 若干強いと考えることにする。
            prod_base =  (1.0 + np.log1p(self.tech_level) * 4.0) / (1.0 + np.log1p(self.culture_level * 0.3))

            # 必需品の目標量 = 民間需要(1000) + 公共事業
            target_basic = 1000.0 + public_demand
            # 金利が高いと「在庫を嫌う」ため、供給が目標を下回るリスクを持たせる
            # target_basic_adj = target_basic * (1.0 - max(0, effective_rate - 0.06) * 0.4)
            # 金利による供給抑制を公共事業分には適用しない(強制支出)
            target_basic_adj = (1000.0 * (1.0 - max(0, effective_rate - 0.06) * 0.4)) + public_demand

            # 必要な労働力（理論値）
            req_labor_basic = target_basic_adj / prod_base

            # 有効労働力（幸福度が低いとストライキや意欲低下で労働力が減る）
            labor_efficiency = 1.0 + 0.3 * np.tanh(self.mass.happiness / 50.0)
            effective_labor_total = self.labor_total * labor_efficiency

            # 労働配分の決定（前回の配分からの急変を抑える慣性）
            alloc_labor_basic = min(req_labor_basic, effective_labor_total)

            # 必需品への労働割り当て (優先)
            # 社会は「文化目標」を達成しようとするが、労働力が足りなければ必需品が優先される
            alloc_labor_basic = min(req_labor_basic, effective_labor_total)
            # 残りが贅沢品へ
            alloc_labor_luxury = max(0, effective_labor_total - alloc_labor_basic)

            supply_basic = alloc_labor_basic * prod_base
            # 金利が高いと必需品以外の生産を控える
            # 債券(利回り)が文化生産の効率をわずかに下げる(保守化)
            supply_luxury = alloc_labor_luxury / (1.0 + policy_rate * 0.5)

            # 6. 所得・分配
            total_output_val = supply_basic * self.price_basic + supply_luxury * self.price_luxury
            # 賃金シェアを固定的に
            self.mass.income = total_output_val * 0.55
            self.rich.income = total_output_val * 0.45

            # 再分配
            transfer = self.rich.income * self.redistribution_rate
            self.rich.income -= transfer
            self.mass.income += transfer
            # income は資産価格により変わったりするので
            # income_ratio には redistribution_rate を使ったほうがよいようだ。
            #income_ratio = self.mass.income / (self.rich.income + 1e-9)
            income_ratio = self.redistribution_rate

            # 7. 資産価値の評価
            # 株式: 将来の選択肢(文化)を強く反映。リスクに非常に敏感。
            valuation_mult_stock = 1.0 / (effective_rate + 0.05)
            # 技術脅威を「文化との乖離」から算出
            tech_threat = max(0, ((self.tech_bias + self.public_work_rate) - 0.5) * 7.0)
            risk_disc = 1.0 / (1.0 + np.sqrt(tech_threat) * 0.3)
            # 急激な乗数を防ぐため log を活用（飽和性を持たせる）
            target_stock = (self.tech_level * 50 + self.culture_level * 50) * (1 + np.log1p(self.culture_level) * 1.0) * valuation_mult_stock * risk_disc * 0.1
            self.stock_market_val = self.stock_market_val * 0.8 + target_stock * 0.2

            # 債券: 弱い株式 + 公共事業の裏付け
            # 金利上昇で価格(単価)は下がるが、需要(ボリューム)は逃避資金で支えられる
            # 債券価値 = (金利による割引) × (安全性への逃避需要)
            bond_discount = 1.0 / (policy_rate + 0.05)
            # 銀行プレミアム（リスク）が上がるほど、安全な債券への需要が強まる
            safety_demand = 1.0 + self.bank_premium * 10.0 + (0.5 if policy_rate > 0.1 else 0.0)
            # 金利に連動するが、文化増幅効果はない
            target_bond = ((self.tech_level * 50 + self.culture_level * 50 + public_demand * 0.5)) * bond_discount * safety_demand * 0.04
            self.bond_market_val = self.bond_market_val * 0.9 + target_bond * 0.1

            # 資産配分 (金利が高いと債券へ)
            total_assets = self.stock_market_val + self.bond_market_val
            self.rich.asset_stock = self.stock_market_val
            self.rich.asset_bond = self.bond_market_val

            # 8. 消費と幸福度
            # 金利コストが価格に転嫁される
            effective_price_magnifier = 1.0
            effective_price_basic = self.price_basic * (1 + effective_price_magnifier * effective_rate)
            effective_price_luxury = self.price_luxury * (1 + effective_price_magnifier * effective_rate)

            avg_asset_total = (self.rich.asset_stock + self.rich.asset_bond + self.mass.asset_stock) / 100.0
            res_r = self.rich.consume(self.rich.income + (self.rich.asset_stock + self.rich.asset_bond)*0.01, effective_price_basic, effective_price_luxury, self.culture_level, tech_threat, avg_asset_total, income_ratio)
            res_m = self.mass.consume(self.mass.income, effective_price_basic, effective_price_luxury, self.culture_level, tech_threat, avg_asset_total, income_ratio)

            # 9. 価格調整
            # 在庫（バッファ）がない状態で需要が供給を上回ると価格が跳ねる
            def adj(p, dem, sup, buf):
                gap = (dem - sup) / (sup + 1e-9)
                sens = 0.25 / (np.log1p(buf) + 0.1)
                return np.clip(p * (1 + np.clip(gap * sens, -0.05, 0.2)), 0, 30)

            self.price_basic = adj(self.price_basic, res_r['basic_demand']+res_m['basic_demand'] + public_demand, supply_basic, self.culture_level)
            self.price_luxury = adj(self.price_luxury, (res_r['luxury_budget']+res_m['luxury_budget'])/self.price_luxury, supply_luxury, self.culture_level)

            effective_price_basic = self.price_basic * (1 + effective_price_magnifier * effective_rate)
            effective_price_luxury = self.price_luxury * (1 + effective_price_magnifier * effective_rate)

            # 10. 記録
            self.history['stock_val'].append(self.stock_market_val)
            self.history['bond_val'].append(self.bond_market_val)
            self.history['happiness_mass'].append(self.mass.happiness)
            self.history['happiness_rich'].append(self.rich.happiness)
            self.history['price_basic'].append(effective_price_basic)
            self.history['price_luxury'].append(effective_price_luxury)
            self.history['demand_basic'].append(res_r['basic_demand']+res_m['basic_demand'] + public_demand)
            self.history['relative_price'].append(self.price_basic / (self.price_luxury + 1e-9))
            self.history['supply_basic'].append(supply_basic)
            self.history['supply_luxury'].append(supply_luxury)
            self.history['labor_share_basic'].append(alloc_labor_basic / effective_labor_total)
            self.history['tech_level'].append(self.tech_level)
            self.history['public_demand_share'].append(public_demand / (target_basic + 1e-9))
            self.history['asset_ratio_stock'].append(self.stock_market_val / (total_assets + 1e-9))
            self.history['policy_rate'].append(policy_rate)
            self.history['bank_premium'].append(self.bank_premium)
            self.history['culture_level'].append(self.culture_level)
            self.history['luxury_rate'].append(supply_luxury / (supply_luxury + supply_basic))
            self.history['inequality_index'].append((self.rich.asset_stock + self.rich.asset_bond) / (self.mass.income * 10 + 1))

def plot_refined_visualization(scenarios, titles, data_keys, cols=2):
    assert len(titles) == len(data_keys)
    rows = (len(titles) + cols - 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=((20/3) * cols, (18/6) * rows))
    colors = plt.cm.tab10(np.linspace(0, 1, len(scenarios)))
    for i, (name, sim) in enumerate(scenarios.items()):
        for idx, key in enumerate(data_keys):
            ax = axes.flat[idx]
            ax.plot(sim.history[key], label=name if idx==0 else "", color=colors[i], lw=2)

    for idx, ax in enumerate(axes.flat):
        if idx < len(titles):
            ax.set_title(titles[idx], fontsize=11, fontweight='bold')
            ax.grid(True, alpha=0.3)
            if idx == 0: ax.legend(loc='upper left', bbox_to_anchor=(0, 1.3), ncol=3, fontsize=9)
    plt.tight_layout()
    plt.show()
````

## 1. 分配の負の効果

まず最初のアイデアは次のようなものだった。

<blockquote>
金持ちから税を取って、分配すると、インフレになる…とか文化が死ぬ…みたいに言われることがあります。この辺を「分析」してみます。

ここには、分配すると、金持ちが、物価を操作してインフレにするんじゃないかという疑いがあるわけです。しかし、経済とは難しいもので、そんなに簡単に操れるものでもありません(AI さん達があれば可能になるのかもしれませんが)。そこにはそれを可能とする条件があるわけです。

なぜそのような「操作」が可能になるか。すれは、まず分配して需要が増える…だけからは説明できません。なぜなら、人々が文化的で、増えた所得で文化的に「新しいもの」を欲し、その供給がはじまるなら、それは単なる経済成長で、インフレが起こると限らないからです。

しかし、そのように増えた需要のうち、一部、供給が限られたものがあるとします。すると、その供給を増やすために、労働力が必要とされます。そういうことが随所で起きて労働力が逼迫してくるようになると、あらゆる生産物で供給が足りなくなり、インフレが起こりえます。ここがまるで物価が操作されたように見えてくる点なのだと思います。

このようなことが起きないようにするために、まず供給を増やすことを考えてみましょう。すると、生産物の技術革新に「分配」して、その(工場供給的)生産性が上がったとします。これで、同様のものを売ってる他者がつぶれると、そのコストが社会にふりかかるだけなので、ここは、その生産性向上分だけ、全体としてそのモノの価格が下がると考えます。ただ、これは名目的な価格下落と観測されることは少なくて、価格上昇の中、実効的なそのモノの価格が下がっていると観測されるでしょう。その分、消費者には余裕が出てきます。

こういうとき何が買われるか。それが分配の仕方に関わります。これはこれまでの議論とは少し離れた議論になります。文化が守られるような分配の仕方はあるのだと思われます。一方、それとは別に、マスに分配することで経済を大きくし、それによって、先の文化が守られるような分配よりも文化が守られる規模の大きな経済というのはありえるのだと思います。しかし、それが、この上の議論のような供給制限があると、インフレになり、結局文化が守られないような実態になりうる。このとき、シュリンクの仕方によっては、ロシアの戦争経済のように文化を守る方向に強く経済を変形させることは可能なのかもしれません。
</blockquote>

これを示すには次のようなシナリオを実行すればよいだろう。理想状態シナリオは文化・技術・公共への投資がバランスが取れているものである。高分配シナリオは、それをより高分配にしたものである。これにより必需品の需要が増加し、ややインフレぎみになっているのがわかる。そのインフレを抑制しようとして、公共投資を技術投資に回してみたのが、技術投資aシナリオ、文化投資を技術投資に回してみたのが と技術投資bシナリオである。残念ながら、どちらもインフレ抑止に成功しなかった。


````python
scenarios = {
    "1. 理想": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.4, public_work_rate=0.2, redistribution_rate=0.2),
    "2. 高分配": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.4, public_work_rate=0.2, redistribution_rate=0.6),
    "3. 技術投資a": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.5, public_work_rate=0.1, redistribution_rate=0.6),
    "4. 技術投資b": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.5, public_work_rate=0.2, redistribution_rate=0.6),
}

for s in scenarios.values(): s.run()
plot_refined_visualization(
    scenarios,
    titles = [
        "大衆の幸福度", "富裕層の幸福度",
        "必需品価格", "贅沢品価格",
        "必需品供給量", "贅沢品供給量",
        "技術レベル", "文化レベル",
    ],
    data_keys = [
        'happiness_mass', 'happiness_rich',
        'price_basic', 'price_luxury',
        'supply_basic', 'supply_luxury',
        'tech_level', 'culture_level',
    ],
)
````

説明すると、高分配による必需品需要増は、consume メソッドの中でほぼハードコーディングされている。高分配にすれば、そうなるように予めプログラムしているのであり、こうなるのはあたり前である。

次に、技術投資がインフレ抑制に役立つはずだとされたのは、技術投資が生産性を向上させ少ない労働力でも必需品がまかなえるようになるはずだからである。

このプログラムでは公共事業は必需品消費にほぼあたる。そして、公共事業と技術投資を除いたものから文化投資がほぼなされるようになっているため、公共事業と技術投資の和を保って動かす技術投資aシナリオでまずはインフレ抑制が観測されるだろうと考えた。

しかし、それが観測されなかったのは、公共事業を削ったため、必需品の総量が小さくなり、その分、価格弾力性的なものが大きくなり、必需品価格が上がりやすくなったためである。

この総量の小ささが必需品価格を高止まりさせるのを確かめるには逆に総量が小さいほど価格変化が起きやすい次の adj のコードを使えばよい。

```
            def adj(p, dem, sup, buf):
                # これにより、規模(sup)が小さいほど変化幅が自然に抑制される
                gap_factor = np.log1p((dem - sup) * sup)

                # 基本感度 (bufによる調整は維持)
                # log1p(sup)が変化幅を大きく持つため、係数は少し調整
                base_sens = 0.01 / (np.log1p(buf) + 0.1)

                # 変化率の計算
                change_rate = gap_factor * base_sens

                # 急激な変化を防ぐためのクリッピング（±20%）
                change_rate = np.clip(change_rate, -0.2, 0.2)

                # 価格更新
                return p * (1 + change_rate)
```
    
では、次に、公共事業を削らなかった技術投資bシナリオでインフレ抑制が失敗したのはなぜか。これは、文化レベルが下がってしまったため、文化が持っている需要バッファ機能の働きが薄れ、必需品価格が上昇してしまったからのようだ。

Gemini さんのこの節の結論はこうなる。

<blockquote>
ロシアの戦争経済が文化を守る方向に経済を変形させ得るのは、それが強力な「供給管理」と「資源の強制割り当て」を伴うからである。しかし、自由意志に基づく経済圏においてインフレを抑えつつ分配を最大化するには、以下の二点が不可欠であることが今回のシミュレーションで示された。

  1. 公共事業の維持: 市場の総量（供給規模）を確保し、価格弾力性を安定させる。

  2. 文化投資の維持: 需要のバッファを機能させ、消費の質的向上を促す。

技術投資単体での解決を図るのではなく、これらの「土台」を維持した上での技術革新こそが、インフレなき高分配社会の鍵となる。
</blockquote>


## 2. 文化と技術の対立

次のアイデアである。

<blockquote>
文化と技術の対立(?)を考えます。

文化が高いと技術も高いと思われます。文化が高いと需要も多いでしょう。

一方、技術が高いからといって文化が高いとは限りません。技術はそのままでは需要を生みません。技術はデフレを生む？ …それも少し違って、技術は軍事大国化により需要を創出することを通じて、人々の生活領域に脅威をもたらし、全体の幸福度を下げるのでしょう。

幸福度とは何でしょうか？ 文化が高いレベルにあるものがあれば、全体に幸福があるとできるかもしれません。しかし、一方で、格差が大きいと幸福度は下がるでしょう。幸福度からはいったん離れます。

『宗教学雑考集』《必需品と贅沢品の宇宙的独立関係》という論考を前に考えました。技術が高いだけというのは、必需品一辺倒になることでしょう。すべてがある意味必需品になります。この含意は、選択の余地がない…ということになります。では選択の余地があることが幸福という論理につなげるべきでしょうか？ いえ、これも急ぎ過ぎです。

選択の余地がある場合、何が有利なのか、それは株式市場の論理とつながります。私は＞債権では社会の富は増えない。株式のみ社会の富を増やせる。＜([cocolog:75599118])…と考えていて、一方、マルクス経済学の剰余価値に関する議論で、在庫(研究費も含む)の増加が別に社会の富を増やすという議論が出てきて、この、株価の増価と在庫の増加が、(銀行の信用創造や債券価格の調整などが間にはさまった上で、) 安定した関係にあることが、安定した経済成長が可能になる条件ではないかと考えたことがありました。

この株式に必要な永続企業の前提が、選択の余地のあるところでは、その選択の両方に成り立つことになるため、そこに株式の富がより多く生じるようになると考えます。これが文化が需要を増やす根本原因なのでしょう。

消費者に個々に選択の余地があることは幸福度が増すという面もあるでしょう。しかし、それは錯覚に近いものかもしれません。別の誰かの持ってる資産に幸福度を見出すのと同じ錯覚なのでしょう。幸福度を直接最適化の基準にしては失敗するのだと思われます。

必需品と究極的には独立な贅沢品＝知的生産物の目標量に達するよう需要と供給を組むとき、それを制約するのが幸福度なのかもしれません。その制約方法が文化なのかも。
</blockquote>

これを示すために技術のみ変化させてみる。


````python
scenarios = {
    "1. 理想": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.4, public_work_rate=0.2, redistribution_rate=0.2),
    "2. 技術重視": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.6, public_work_rate=0.2, redistribution_rate=0.2),
    "3. 技術軽視": ComplexEconomySimulation(rate_scenario="low_stable", tech_bias=0.2, public_work_rate=0.2, redistribution_rate=0.2),
}

for s in scenarios.values(): s.run()
plot_refined_visualization(
    scenarios,
    titles = [
        "不平等度", "贅沢品割合",
        "大衆の幸福度", "株式市場価値",
    ],
    data_keys = [
        'inequality_index', 'luxury_rate',
        'bank_premium', 'stock_val',
    ],
)
````

幸福度は、技術重視→理想→技術軽視の順に高く、不平等度はその逆順で、なんとも言えない。ただ、幸福度は consume 関数をご覧になればわかるように一定程度不平等度も考慮しているので、理想がもっとも幸福度が高いわけではない…のは都合が悪いかもしれない。ただそのあたり、幸福度を最適化しているわけではない＝理想としない…というところで、アイデア通りであるとは言える。

ちなみに consume 関数の幸福度は、必需品が必要量あり、選択の自由があって贅沢品が買え、保有資産から不平等さを感じなくてすみ、技術優先の全体国家的でない…場合に高くなるようになっている。

そして、注目すべきは贅沢品割合で、これが宇宙的独立で、シミュレーションに与えるパラメータでだいたい決まってくる感じになっている。逆に言えば、どれぐらい贅沢品＝知的財産を作りたいかで、国家体制が決まってくるということだ。

Gemini さんのこの節の結論はこうなる。

<blockquote>
贅沢品（知的生産物）の割合は、単なる市場の結果ではなく、その社会がどのような「知的フロンティア」を目指すかというパラメータによって決定される。

  * 高贅沢品割合: 文化と選択の自由を重視する、株式駆動型の自由経済。

  * 低贅沢品割合: 技術と必需品効率を重視する、債権・軍事駆動型の全体主義経済。

我々がどのシナリオを選択するかは、経済学的な正解の問題ではなく、どのような文明を築きたいかという政治哲学的選択の問題である。
</blockquote>


## 3. 金利と物価

次のアイデアである。

<blockquote>
金利について考えます。

政策金利とそれに銀行プレミアムを乗せた銀行(貸出)金利があるとします。

金利によって、株式から得られる利益から逆算して、その株式の資産価値が決まります。それによって適切な在庫の量が決まります。このときに問題となるのは政策金利でしょう。

複数の企業が生産するとき、一つの企業が技術革新で低コストで作れるようになった場合、従来企業の赤字は、「政府」的なものといったんなり、それが新企業の利益と相殺される…これがある種の政策的分配になります。この分配を担うのが銀行機能で、このような赤字の発生の度合いによって銀行プレミアムは上下すると考えることができるでしょう。銀行プレミアム上昇分は消費者に転嫁されます。

これが先の「名目的な価格下落と観測されることは少なくて、価格上昇の中、実効的なそのモノの価格が下がっていると観測される」と接続します。インフレ基調の中、株式の資産価値はその分減価し、それによって許される在庫量が変化します。このとき、銀行プレミアムがあることで生産されない財も出てくるのでしょう。(このあたりの(心の中の)モデル化はややあやしいです。)

非必需品経済で「選択の余地」がある場合、富が増えていますから、赤字倒産も少なくなり、必需品経済に比べて銀行プレミアムが少なく済むということが考えられます。

一方、供給危機にある場合は、銀行は必ず儲けることができるため、必需品経済でも銀行プレミアムを小さくできます。しかし、全体としての生産を促すため、非必需品経済では在庫がない分「選択の余地」を削りながら、政策金利を上げて全体在庫を調整せざるを得なくなるのでしょう。「選択の余地」を削れば、やがて銀行プレミアムも増やさざるを得ず、悪循環に陥るのかもしれません。
</blockquote>

今回のシミュレーションを通じて得られた最大の知見は、「文化（知的・多様的な遊び）」が、単なる贅沢ではなく、経済システム全体の破綻を防ぐ「動的なバッファ（緩衝材）」として機能しているという点です。
